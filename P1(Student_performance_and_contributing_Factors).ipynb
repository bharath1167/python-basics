# Question-1: How does the number of hours studied per week correlate with the final exam score?

```python
a.isna().sum()
```

```python
correlation = a['Hours_Studied'].corr(a['Exam_Score'])
print(correlation)
```

```python
# Correlation using heatmap
b = a['Hours_Studied'].corr(a['Exam_Score'])
b = pd.DataFrame([[b]], columns=['Correlation'])
sns.heatmap(b, annot=True)
```
