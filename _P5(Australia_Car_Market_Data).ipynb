#Question-1:What is the average price of cars across different brands, and which brands have the highest and lowest average prices?

```python
a.groupby('Brand')['Price'].mean()
```

#Question-2:How does the condition of a car (Status) affect its price?

```python
a.groupby('Status')['Price'].mean()
```

#Question-3: Is there a correlation between mileage (Kilometers) and the price of a car?

```python
# Is there a correlation between mileage (Kilometers) and the price of a car
b = a['Kilometers'].corr(a['Price'])
b = pd.DataFrame([[b]], columns=['Correlation Between Kilometers and Price'])
sns.heatmap(b, annot=True)
```
